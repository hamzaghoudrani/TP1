```python
# =============================================================================
# EXERCISE 3 — House Price Prediction with Deep Learning (Adam Optimizer)
# Dataset : AmesHousing.csv
# Goal    : Predict the sale price of a house from its features
# Changes : Switched optimizer to Adam, added improvements for better results
# =============================================================================

# ── 1. IMPORTS ────────────────────────────────────────────────────────────────
import numpy as np                         # numerical operations & array handling
import matplotlib.pyplot as plt            # plotting graphs
import pandas as pd                        # data manipulation (DataFrames)
import seaborn as sns                      # statistical data visualization
import tensorflow as tf                    # deep learning framework
from tensorflow import keras               # high-level Keras API inside TensorFlow
from tensorflow.keras import layers        # individual neural-network layer types
from tensorflow.keras.callbacks import (   # training callbacks for better control
    EarlyStopping,                         #   stop training when validation stops improving
    ReduceLROnPlateau                       #   reduce LR automatically when stuck
)

# Print TensorFlow version so results are reproducible and debug-friendly
print("TensorFlow version:", tf.__version__)

# Fix global random seed so every run gives the same weights & splits
tf.random.set_seed(42)
np.random.seed(42)


# ── 2. LOAD DATASET ───────────────────────────────────────────────────────────
df = pd.read_csv('AmesHousing.csv')       # load the Ames Housing CSV into a DataFrame

print("Dataset shape:", df.shape)         # (rows, columns) — useful sanity check
print("\nFirst 5 rows:")
print(df.head())                           # preview the raw data
print("\nDescriptive statistics:")
print(df.describe())                       # mean / std / min / max for every numeric column


# ── 3. HANDLE MISSING VALUES ──────────────────────────────────────────────────

# --- 3a. Audit: count & percentage of NaN per column ---
VM = pd.DataFrame({
    'Column'    : df.columns.values,
    'Nb missing': df.isna().sum().values,
    '% missing' : 100 * df.isna().sum().values / len(df),  # convert to percentage
})
VM = VM[VM['Nb missing'] > 0]             # keep only columns that actually have NaNs
print("\nColumns with missing values (sorted):")
print(VM.sort_values(by='Nb missing', ascending=False).reset_index(drop=True))

# --- 3b. Drop columns missing more than 40 % of their values ---
# Imputing columns that are mostly empty would inject too much fake data
threshold = 40                            # drop threshold in percent
cols_to_drop = VM[VM['% missing'] > threshold]['Column'].tolist()
df = df.drop(columns=cols_to_drop)        # remove those unreliable columns
print(f"\n{len(cols_to_drop)} columns dropped (>{ threshold}% missing):", cols_to_drop)

# --- 3c. Inspect what remains so we choose the right fill strategy ---
cols_with_mv = df.columns[df.isna().sum() > 0]  # columns that still have NaNs
print("\nRemaining columns with missing values:")
for col in cols_with_mv:
    print(f"  {col}  →  samples: {df[col].unique()[:5]}")

# --- 3d. Fill numeric columns with their column mean ---
# Mean is a neutral estimate that keeps the overall distribution intact
num_MV = ['Lot Frontage', 'Mas Vnr Area', 'BsmtFin SF 1', 'BsmtFin SF 2',
          'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath',
          'Bsmt Half Bath', 'Garage Yr Blt', 'Garage Cars', 'Garage Area']

for col in num_MV:
    if col in df.columns:                 # skip if the column was already dropped
        mean_val = df[col].mean()
        df[col] = df[col].fillna(mean_val)
        print(f"  [NUM] '{col}' → filled with mean {mean_val:.2f}")

# --- 3e. Fill categorical columns with their column mode (most frequent value) ---
cat_MV = [x for x in cols_with_mv if x not in num_MV]  # everything not numeric
for col in cat_MV:
    if col in df.columns:
        mode_val = df[col].mode().to_numpy()[0]           # most common category
        df[col] = df[col].fillna(mode_val)
        print(f"  [CAT] '{col}' → filled with mode '{mode_val}'")

# Verify: no NaN must remain before entering the model
print("\nTotal missing values remaining:", df.isna().sum().sum())  # must be 0


# ── 4. ENCODE CATEGORICAL VARIABLES ──────────────────────────────────────────

# MS SubClass is stored as int by pandas but is actually a nominal category code
df['MS SubClass'] = df['MS SubClass'].astype(str)  # convert to string so get_dummies picks it up

# Select 20 meaningful features to keep the encoded dataset manageable
# Using all 82 raw columns would create 200+ binary columns → higher overfitting risk
selected_features = [
    'MS SubClass', 'MS Zoning',
    'Lot Frontage', 'Lot Area',
    'Neighborhood',
    'Overall Qual', 'Overall Cond',
    'Year Built',
    'Total Bsmt SF', '1st Flr SF', '2nd Flr SF',
    'Gr Liv Area',
    'Full Bath', 'Half Bath',
    'Bedroom AbvGr', 'Kitchen AbvGr', 'TotRms AbvGrd',
    'Garage Area', 'Pool Area',
    'SalePrice'                                   # target — kept here, popped later
]
df = df[selected_features]                        # trim to the 20 selected columns
df = pd.get_dummies(df)                           # one-hot-encode every remaining string column
print("\nColumns after One-Hot Encoding:", df.shape[1])  # ~67 columns expected


# ── 5. TRAIN / TEST SPLIT ─────────────────────────────────────────────────────
# random_state=9 guarantees the same split every run (reproducibility)
train = df.sample(frac=0.8, random_state=9)       # 80 % for training
test  = df.drop(train.index)                       # 20 % for testing (never seen during training)

# Separate target from features — pop() removes the column and returns it as a Series
train_target = train.pop('SalePrice')
test_target  = test.pop('SalePrice')

print(f"\nTrain : {train.shape[0]} rows | Test : {test.shape[0]} rows")
print(f"Train SalePrice range : ${train_target.min():,.0f} – ${train_target.max():,.0f}")
print(f"Test  SalePrice range : ${test_target.min():,.0f}  – ${test_target.max():,.0f}")


# ── 6. STANDARDIZATION (Z-SCORE) ─────────────────────────────────────────────
# Without scaling, features with large magnitudes (e.g. Lot Area ~10 000)
# dominate gradient updates while small features (e.g. Full Bath ~1–3) are ignored.
# Z-score puts every feature on the same scale: mean=0, std=1.
# IMPORTANT: we compute mean & std on TRAIN only and apply the SAME transform to TEST
# to avoid data leakage (the model must never "see" test statistics).

for col in train.columns:
    col_mean = train[col].mean()           # training mean for this feature
    col_std  = train[col].std()            # training std for this feature

    if col_std == 0:                       # guard: constant column → avoid division by zero
        col_std = 1e-20

    train[col] = (train[col] - col_mean) / col_std   # standardize training set
    test[col]  = (test[col]  - col_mean) / col_std   # apply SAME params to test set

print(f"\nAfter standardization — '{train.columns[0]}':")
print(f"  Train mean ≈ {train[train.columns[0]].mean():.6f}  (should be ~0)")
print(f"  Train std  ≈ {train[train.columns[0]].std():.6f}   (should be ~1)")


# ── 7. BUILD THE NEURAL NETWORK ───────────────────────────────────────────────
# keras.Sequential stacks layers in order; data flows straight from input to output.
# Dense = fully-connected layer (each neuron connected to all neurons of the previous layer).
# Activations: ReLU and Swish alternate — Swish tends to give smoother gradients in deep nets.
# Dropout(0.3): randomly zeros 30 % of neurons each training step → reduces overfitting.
# BatchNormalization: normalises activations inside the network → faster, more stable training.
# Output layer: 1 neuron, no activation (linear) → can predict any real price value.

model = keras.Sequential([
    # Input layer: must know the number of input features
    layers.Dense(128, activation='relu', input_shape=[train.shape[1]]),
    layers.BatchNormalization(),           # stabilise & accelerate learning right from the start
    layers.Dropout(0.3, seed=42),          # drop 30 % of neurons to combat overfitting

    layers.Dense(128, activation='swish'), # hidden layer 1 — Swish is smooth and differentiable
    layers.BatchNormalization(),
    layers.Dropout(0.2, seed=42),          # lighter dropout in later layers

    layers.Dense(64, activation='relu'),   # hidden layer 2 — ReLU is fast and simple
    layers.BatchNormalization(),

    layers.Dense(64, activation='swish'),  # hidden layer 3
    layers.BatchNormalization(),

    layers.Dense(32, activation='relu'),   # hidden layer 4 — gradually narrow the network
    layers.Dense(1)                        # output layer: one predicted sale price (linear)
])

model.summary()                            # print architecture table with parameter counts


# ── 8. COMPILE THE MODEL ──────────────────────────────────────────────────────
# Loss   : MSE penalises large errors heavily — good for price prediction
# Optimizer: Adam combines momentum + adaptive per-parameter learning rates
#            → converges faster and more reliably than plain SGD or RMSprop
# Metric : MAE is easy to interpret ("on average, off by $X")

optimizer = tf.keras.optimizers.Adam(
    learning_rate=0.001                    # 1e-3 is the standard Adam starting point
)

model.compile(
    loss      = tf.keras.losses.MeanSquaredError(),
    optimizer = optimizer,
    metrics   = ['mae']
)


# ── 9. CALLBACKS ──────────────────────────────────────────────────────────────
# EarlyStopping: halt training when val_mae stops improving for 15 epochs
#   restore_best_weights=True → revert to the best checkpoint automatically
early_stop = EarlyStopping(
    monitor='val_mae',
    patience=15,                           # wait 15 epochs before stopping
    restore_best_weights=True,
    verbose=1
)

# ReduceLROnPlateau: if val_mae stagnates for 7 epochs, divide LR by 2
#   helps the optimizer escape flat regions without manual tuning
reduce_lr = ReduceLROnPlateau(
    monitor='val_mae',
    factor=0.5,                            # new_lr = old_lr * 0.5
    patience=7,
    min_lr=1e-6,                           # do not go below this floor
    verbose=1
)


# ── 10. TRAIN THE MODEL ───────────────────────────────────────────────────────
# epochs=150: upper limit — EarlyStopping will stop sooner if needed
# validation_split=0.2: 20 % of train data is used to monitor generalisation each epoch
# callbacks: EarlyStopping + ReduceLROnPlateau applied during training

history = model.fit(
    train, train_target,
    epochs           = 150,
    validation_split = 0.2,
    callbacks        = [early_stop, reduce_lr],
    verbose          = 1
)


# ── 11. PLOT TRAINING CURVES ──────────────────────────────────────────────────
hist_df  = pd.DataFrame(history.history)  # convert training log to DataFrame
hist_df['epoch'] = history.epoch          # add epoch numbers as a column
n_epochs = hist_df.shape[0]               # actual number of epochs run (may be < 150)

fig, ax = plt.subplots(figsize=(14, 8))
ax.plot(np.arange(n_epochs), hist_df['mae'],     label='Training MAE',   lw=3, color='red')
ax.plot(np.arange(n_epochs), hist_df['val_mae'], label='Validation MAE', lw=3, color='blue')
ax.set_xlabel('Epoch')
ax.set_ylabel('MAE (Mean Absolute Error — $)')
ax.set_title('Training vs Validation MAE over Epochs')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('training_curves.png', dpi=150)   # save figure to disk
plt.show()


# ── 12. EVALUATE ON TEST SET ──────────────────────────────────────────────────
# This is the final, unbiased measure of model quality on data never seen during training
test_loss, test_mae = model.evaluate(test, test_target, verbose=0)
print(f"\nTest MSE : {test_loss:,.2f}")
print(f"Test MAE : ${test_mae:,.2f}")
print(f"→ On average the model is off by ${test_mae:,.0f} on unseen house prices.")


# ── 13. SINGLE-SAMPLE PREDICTION ─────────────────────────────────────────────
# Pick one house from the test set and compare prediction vs reality
sample     = test.iloc[[10]]                       # select the 11th row as a 1-row DataFrame
prediction = model.predict(sample).squeeze()       # squeeze (1,1) array → scalar
actual     = test_target.iloc[10]                  # ground-truth sale price

print("\nPrediction example (house #10 in test set):")
print(f"  Model prediction : ${prediction:,.2f}")
print(f"  Actual price     : ${actual:,.2f}")
print(f"  Difference       : ${abs(prediction - actual):,.2f}")


# ── 14. PREDICTED vs ACTUAL SCATTER PLOT ─────────────────────────────────────
# Each dot = one house in the test set.
# If dots cluster around the red diagonal (y = x) the model is accurate.
all_predictions = model.predict(test).squeeze()   # predict every house in the test set

plt.figure(figsize=(10, 8))
plt.scatter(test_target, all_predictions,          # x = actual, y = predicted
            alpha=0.4, color='steelblue', s=40, label='Predicted price')
plt.plot([test_target.min(), test_target.max()],   # draw perfect-prediction diagonal
         [test_target.min(), test_target.max()],
         color='red', lw=2, linestyle='--', label='Perfect prediction (y = x)')
plt.xlabel('Actual Sale Price ($)')
plt.ylabel('Predicted Sale Price ($)')
plt.title('Predicted vs Actual House Prices — Test Set')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('predicted_vs_actual.png', dpi=150)   # save figure to disk
plt.show()

# ── 15. RESIDUAL DISTRIBUTION ─────────────────────────────────────────────────
# Residuals = actual − predicted.  A model is well-calibrated when residuals
# are centred at 0 and roughly bell-shaped (no systematic over/under-estimation).
residuals = test_target.values - all_predictions

plt.figure(figsize=(10, 5))
sns.histplot(residuals, kde=True, color='steelblue', bins=40)  # histogram + KDE curve
plt.axvline(0, color='red', linestyle='--', lw=2, label='Zero error')
plt.xlabel('Residual (Actual − Predicted) in $')
plt.ylabel('Count')
plt.title('Residual Distribution — Test Set')
plt.legend()
plt.tight_layout()
plt.savefig('residuals.png', dpi=150)
plt.show()

print("\nDone. All plots saved.")
```
